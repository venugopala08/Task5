```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style="whitegrid")
%matplotlib inline

# Load the Titanic dataset
df = pd.read_csv('titanic.csv')

# Check the shape and head
print("Shape:", df.shape)
df.head()
```

```python
# Describe numerical features
print("\nDescriptive Statistics:")
display(df.describe())

# Dataset Information
print("\nDataset Info:")
df.info()

# Value counts for categorical columns
print("\nValue Counts for Categorical Columns:")
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col}:\n", df[col].value_counts())
```

```python
# Heatmap (correlation between numerical columns only)
numeric_cols = df.select_dtypes(include='number')

plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot (relationships between features)
sns.pairplot(df, hue='Survived', vars=['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'])
plt.suptitle('Pairplot of Selected Features by Survived', y=1.02)
plt.show()
```

```python
# Survival rate vs Pclass
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate vs Passenger Class")
plt.show()

# Survival rate vs Sex
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate vs Gender")
plt.show()

# Survival rate vs Embarked
sns.barplot(x="Embarked", y="Survived", data=df)
plt.title("Survival Rate vs Embarked Port")
plt.show()
```

```python
df.hist(figsize=(15,10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Columns", fontsize=20)
plt.show()
```

```python
for col in ['Age', 'Fare']:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Survived', y=col, data=df)
    plt.title(f"Boxplot of {col} vs Survived")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Scatterplot of Age vs Fare colored by Survived")
plt.show()
```
